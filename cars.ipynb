```python
df.rename(columns=lambda x: x.strip(), inplace=True)
```

```python
df["cubicinches"] = pd.to_numeric(df["cubicinches"], errors="coerce")
df["weightlbs"] = pd.to_numeric(df["weightlbs"], errors="coerce")
```

```python
#1. Improvement in mpg over the years
plt.figure(figsize=(10, 5))
sns.lineplot(x=df["year"], y=df["mpg"], estimator="mean", ci=None)
plt.xlabel("Year")
plt.ylabel("Average MPG")
plt.title("MPG Trend Over the Years")
plt.show()
```

```python
#2. affect of cubicinches to reach 60 miles per hour
plt.figure(figsize=(10, 5))
sns.scatterplot(x=df["cubicinches"], y=df["time-to-60"])
plt.xlabel("Cubic Inches")
plt.ylabel("Time to 60 mph (seconds)")
plt.title("Effect of Engine Size on Acceleration")
plt.show()
```

```python
#3. Does a specific country prefer vehicles with more horsepower?
plt.figure(figsize=(10, 5))
sns.boxplot(x=df["brand"], y=df["hp"])
plt.xlabel("Country of Origin")
plt.ylabel("Horsepower")
plt.title("Horsepower Preference by Country")
plt.show()
```

```python
#4. how does horse power affect average?
plt.figure(figsize=(10, 5))
sns.scatterplot(x=df["hp"], y=df["mpg"])
plt.xlabel("Horsepower")
plt.ylabel("MPG")
plt.title("Horsepower vs. Fuel Efficiency")
plt.show()
```

```python
#5. how does cubicinches affects average?
plt.figure(figsize=(10, 5))
sns.scatterplot(x=df["cubicinches"], y=df["mpg"])
plt.xlabel("Cubic Inches")
plt.ylabel("MPG")
plt.title("Engine Size vs. Fuel Efficiency")
plt.show()
```

```python
#6. effect of cylinders of mpg
plt.figure(figsize=(10, 5))
sns.boxplot(x=df["cylinders"], y=df["mpg"])
plt.xlabel("Number of Cylinders")
plt.ylabel("MPG")
plt.title("Cylinders vs. Fuel Efficiency")
plt.show()
```

```python
#7. Relation between cubic inches , number of cylinders , horsepower
sns.pairplot(df, vars=["cubicinches", "cylinders", "hp"], kind="reg")
plt.show()
```

```python
#8. Proportion of vehicles with 8 cylinders
cyl_8_proportion = (df[df["cylinders"] == 8].shape[0] / df.shape[0]) * 100
print(f"Proportion of vehicles with 8 cylinders: {cyl_8_proportion:.2f}%")
```

```python

# 9. Direct proportion between cubic inches, cylinders, and horsepower
sns.heatmap(df[["cubicinches", "cylinders", "hp"]].corr(), annot=True, cmap="coolwarm")
plt.title("Correlation between Cubic Inches, Cylinders, and Horsepower")
plt.show()
```

```python
#10. Inverse proportion between horsepower and MPG
correlation = df["hp"].corr(df["mpg"])
print(f"Correlation between Horsepower and MPG: {correlation:.2f}")
```
